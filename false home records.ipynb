## points earned at home/away

```python
homepoints  = []
awaypoints  = []
fhomepoints = []
totalpoints = []
thpoints    = []

for cl, std, city, dom in zip (hbdf['club'], hbdf['homebase'], hbdf['city'], hbdf['dom']):
    
    if 'Madura' in cl:
        hpe  = df.loc[( (df['home']==cl) & ( (df['stadium']==std)|(df['city']==dom) ) )]['hpoint'].sum()
        fhpe = df.loc[( (df['home']==cl) & ( (df['stadium']!=std)&(df['city']!=dom) ) )]['hpoint'].sum()
        
    else:
        hpe  = df.loc[( (df['home']==cl) & ( (df['stadium']==std)|(city==dom) ) )]['hpoint'].sum()
        fhpe = df.loc[( (df['home']==cl) & ( (df['stadium']!=std)&(city!=dom) ) )]['hpoint'].sum()
            
    ape  = df.loc[(df['away']==cl)]['apoint'].sum()
    
    homepoints.append(hpe)
    fhomepoints.append(fhpe)
    awaypoints.append(ape)
    
    thpoints.append(hpe+fhpe)
    
    totalpoints.append(hpe+fhpe+ape)
```

```python
klsmn = pd.DataFrame ( {'Club': hbdf['club'],
                        'Home Points': homepoints,
                        'False Home Points': fhomepoints,
                        'Away Points': awaypoints,
                        'Total Points': totalpoints,
                        'th': thpoints
})
klsmn
```

## points lost at home/away

```python
def itung (syrt):
    earned = syrt.sum()#df.loc[syrt]['hpoint'].sum()
    fulll = syrt.count()*3#df.loc[syrt]['hpoint'].count()*3
    lost = fulll-earned
    return (lost)
```

```python
homelost  = []
awaylost  = []
fhomelost = []
totallost = []
thlost    = []
for cl, std, city, dom in zip (hbdf['club'], hbdf['homebase'], hbdf['city'], hbdf['dom']):
    
    if 'Madura' in cl or city==dom:
        print('if',cl)
        # points earned
        #hpe  = df.loc[( (df['home']==cl) & ( (df['stadium']==std)|(df['city']==dom) ) )]['hpoint'].sum()
        #fhpe = df.loc[( (df['home']==cl) & ( (df['stadium']!=std)&(df['city']!=dom) ) )]['hpoint'].sum()
        # total played
        #hpfull  = df.loc[( (df['home']==cl) & ( (df['stadium']==std)|(df['city']==dom) ) )]['hpoint'].count()*3
        #fhpfull = df.loc[( (df['home']==cl) & ( (df['stadium']!=std)&(df['city']!=dom) ) )]['hpoint'].count()*3
        # points lost
        #hplost  = hpfull-hpe
        #fhplost = fhpfull-fhpe
        hplost = itung(df.loc[( (df['home']==cl) & ( (df['stadium']==std)|(df['city']==dom) ) )]['hpoint'])
        fhplost = itung(df.loc[( (df['home']==cl) & ( (df['stadium']!=std)&(df['city']!=dom) ) )]['hpoint'])
        
    else:
        print('else',cl)
        # points earned
        #hpe  = df.loc[( (df['home']==cl) & (df['stadium']==std) )]['hpoint'].sum()
        #fhpe = df.loc[( (df['home']==cl) & (df['stadium']!=std) )]['hpoint'].sum()
        #print('point earned', fhpe)
        # total played
        #hpfull  = df.loc[( (df['home']==cl) & (df['stadium']==std) )]['hpoint'].count()*3
        #fhpfull = df.loc[( (df['home']==cl) & (df['stadium']!=std) )]['hpoint'].count()*3
        #print('point full', fhpfull)
        # points lost
        #hplost  = hpfull-hpe
        #fhplost = fhpfull-fhpe
        #print('point lost', fhplost)
        hplost  = itung(df.loc[( (df['home']==cl) & (df['stadium']==std) )]['hpoint'])
        fhplost = itung(df.loc[( (df['home']==cl) & (df['stadium']!=std) )]['hpoint'])
            
    #ape  = df.loc[(df['away']==cl)]['apoint'].sum()
    #apfull = df.loc[(df['away']==cl)]['apoint'].count()*3
    aplost = itung(df.loc[(df['away']==cl)]['apoint'])
    
    homelost.append(hplost)
    fhomelost.append(fhplost)
    awaylost.append(aplost)
    
    thlost.append(hplost+fhplost)
    
    totallost.append(hplost+fhplost+aplost)
```

```python
lostpoints = pd.DataFrame ( {'Club': hbdf['club'],
                        'Home lost': homelost,
                        'False Home lost': fhomelost,
                        'Away lost': awaylost,
                        'Total lost': totallost,
                        'th lost': thlost
})
lostpoints
```
